This didn't work because the americas gdp data has an extra column "continent" that is filled with "Americas" values.

In order to fix this we can use pandas [DataFrame.drop](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html) like this:

`americas_gdp_data.drop('continent', axis=1)`

```python
incontinent_americas_gdp_data = americas_gdp_data.drop('continent', axis=1)
incontinent_americas_gdp_data.mean()
```

```python
europe_gdp_data.mean()
```

```python
mask_higher = europe_gdp_data > europe_gdp_data.mean()
```

```python
print(mask_higher)
```

```python
wealth_score = mask_higher.aggregate('sum', axis=1) / len(europe_gdp_data.columns)
```

```python
print(wealth_score)
```

```python
europe_gdp_data.groupby(wealth_score).sum()
```
